### Problem 5: Engineering Features

Now that you understand your business objective, we will build a basic model to get started.  Before we can do this, we must work to encode the data.  Using just the bank information features, prepare the features and target column for modeling with appropriate encoding and transformations.

````python
# Separate the data into the X, y datasets
X, y = df.drop('y', axis=1), df[['y']]
display(X.head())
display(y.value_counts())
display(np.unique(y))
````

````python
# Recode Y
y['y'] = np.where(y['y'] == 'yes', 1, 0)
````

````python
# Encode the descriptive features
ct = make_column_transformer((StandardScaler(), make_column_selector(dtype_include=np.number)),
                             (OneHotEncoder(drop='first'), make_column_selector(dtype_include=object)))
X = pd.DataFrame(ct.fit_transform(X), columns=ct.get_feature_names_out())
X.head()
````

### Problem 6: Train/Test Split

With your data prepared, split it into a train and test set.

````python
# Create a test train split
X_train, X_test, y_train, y_test = train_test_split(X, y, stratify=y)
````

### Problem 7: A Baseline Model

Before we build our first model, we want to establish a baseline.  What is the baseline performance that our classifier should aim to beat?

````python
# Build and fit a dummy classifier
start_time = time.time()
dummy_clf = DummyClassifier(strategy='most_frequent').fit(X_train, y_train)
end_time = time.time()
dummy_time = (end_time - start_time)/60

# Calc predictions
dummy_test_preds = dummy_clf.predict(X_test)
dummy_train_preds = dummy_clf.predict(X_train)

# Display Confusion Matrix
ConfusionMatrixDisplay(confusion_matrix(y_test, dummy_test_preds)).plot()
plt.suptitle('Dummy Classifier Confusion Matrix')

# Print the classification report
print(classification_report(y_test, dummy_test_preds))

# Intialize Result data frame
results_df = pd.DataFrame({'Model': ['Baseline'],
                           'Train Time': [f'{dummy_time:.3f}m'],
                           'Train Accuracy': [accuracy_score(y_train, dummy_train_preds)],
                           'Train Recall': [recall_score(y_train, dummy_train_preds)],
                           'Test Accuracy': [accuracy_score(y_test, dummy_test_preds)],
                           'Test Recall': [recall_score(y_test, dummy_test_preds)]})

display(results_df)
````

### Problem 8: A Simple Model

Use Logistic Regression to build a basic model on your data.  

````python
# Initialize Estimators
standard_scaler = StandardScaler()
lr_clf = LogisticRegression(n_jobs=-1)

# Fit a pipeline
  
lr_pipe = Pipeline([
    ('standard_scaler', standard_scaler),
    ('logistic_regression', lr_clf)
])

# Fit the Logistic regression
start_time = time.time() 
lr_pipe.fit(X_train, y_train)
end_time = time.time()
lr_time = (end_time - start_time)/60

# Save predictions
lr_test_preds = lr_pipe.predict(X_test)
lr_train_preds = lr_pipe.predict(X_train)
````

### Problem 9: Score the Model

What is the accuracy of your model?

````python
# Display Confusion Matrix
ConfusionMatrixDisplay(confusion_matrix(y_test, lr_test_preds)).plot()
plt.suptitle('Logistic Regression Classifier Confusion Matrix')

# Print the classification report
print(classification_report(y_test, lr_test_preds))

# Intialize Result data frame
lr_results_df = pd.DataFrame({'Model': ['Logistic Regression'],
                              'Train Time': [f'{lr_time:.3f}m'],
                              'Train Accuracy': [accuracy_score(y_train, lr_train_preds)],
                              'Train Recall': [recall_score(y_train, lr_train_preds)],
                              'Test Accuracy': [accuracy_score(y_test, lr_test_preds)],
                              'Test Recall': [recall_score(y_test, lr_test_preds)]})

results_df = pd.concat([results_df, lr_results_df])

display(results_df)
````

### Problem 11: Improving the Model

Now that we have some basic models on the board, we want to try to improve these.  Below, we list a few things to explore in this pursuit.

- More feature engineering and exploration.  For example, should we keep the gender feature?  Why or why not?
- Hyperparameter tuning and grid search.  All of our models have additional hyperparameters to tune and explore.  For example the number of neighbors in KNN or the maximum depth of a Decision Tree.  
- Adjust your performance metric

If there were a feature denoting gender it should be dropped so that the model does
not discriminate.

````python
# Initialize estimators
knn_clf = KNeighborsClassifier()
poly_features = PolynomialFeatures(degree=2)
standard_scaler = StandardScaler()
param_grid = {'knn_clf__n_neighbors': [1, 3, 5 , 7]}

# Build Pipeline
knn_pipe = Pipeline([('poly_features', poly_features),
                     ('standard_scaler', standard_scaler),
                     ('knn_clf', knn_clf)])

# Execute grid search
knn_grid = GridSearchCV(knn_pipe, param_grid=param_grid, n_jobs=-1)
start_time = time.time()
knn_grid.fit(X_train, y_train.to_numpy().ravel())
end_time = time.time()
knn_time = (end_time - start_time)/60

# Save the preditions
knn_test_preds = knn_grid.predict(X_test)
knn_train_preds = knn_grid.predict(X_train)

# Display Confusion Matrix
ConfusionMatrixDisplay(confusion_matrix(y_test, knn_test_preds)).plot()
plt.suptitle('Logistic Regression Classifier Confusion Matrix')

# Print the classification report
print('Training Classification Report ****************')
print(classification_report(y_train, knn_train_preds))
print('Test Classification Report ********************')
print(classification_report(y_test, knn_test_preds))

# Intialize Result data frame
knn_results_df = pd.DataFrame({'Model': ['KNN'],
                               'Train Time': [f'{knn_time:.3f}m'],
                               'Train Accuracy': [accuracy_score(y_train, knn_train_preds)],
                               'Train Recall': [recall_score(y_train, knn_train_preds)],
                               'Test Accuracy': [accuracy_score(y_test, knn_test_preds)],
                               'Test Recall': [recall_score(y_test, knn_test_preds)]})

results_df = pd.concat([results_df, knn_results_df])

display(results_df)
````

````python
# Initialize Estimators
standard_scaler = StandardScaler()
lro_clf = LogisticRegression(max_iter=10_000)
param_grid = {'lr_clf__C':[.001, .01, 1, 10, 100, 1000]}

# Fit a pipeline    
lro_pipe = Pipeline([
    ('standard_scaler', standard_scaler),
    ('lr_clf', lr_clf)
])

# Execute grid search
lro_grid = GridSearchCV(lro_pipe, param_grid=param_grid, n_jobs=-1)
start_time = time.time()
lro_grid.fit(X_train, y_train.to_numpy().ravel())
end_time = time.time()
lro_time = (end_time - start_time)/60

# Get predictions
lro_test_preds = lro_grid.predict(X_test)
lro_train_preds = lro_grid.predict(X_train)

# Display Confusion Matrix
ConfusionMatrixDisplay(confusion_matrix(y_test, lr_test_preds)).plot()
plt.suptitle('Logistic Reg. Optim. Classifier Confusion Matrix')

# Print the classification report
print('Training Classification Report ****************')
print(classification_report(y_train, lro_train_preds))
print('Test Classification Report ********************')
print(classification_report(y_test, lro_test_preds))

# Intialize Result data frame
lro_results_df = pd.DataFrame({'Model': ['Logistic Optimized'],
                               'Train Time': [f'{lro_time:.3f}m'],
                               'Train Accuracy': [accuracy_score(y_train, lro_train_preds)],
                               'Train Recall': [recall_score(y_train, lro_train_preds)],
                               'Test Accuracy': [accuracy_score(y_test, lro_test_preds)],
                               'Test Recall': [recall_score(y_test, lro_test_preds)]})

results_df = pd.concat([results_df, lro_results_df])

display(results_df)
````

````python
# Initialize Estimators
standard_scaler = StandardScaler()
dt_clf = DecisionTreeClassifier()
param_grid = {'dt_clf__min_impurity_decrease': [0.001, 0.01, 0.02, 0.03, 0.05],
              'dt_clf__max_depth': [2, 5, 10, 20],
              'dt_clf__max_features': ['sqrt', 'log2'],
              'dt_clf__min_samples_split': [0.1, 0.2, 0.05],
              'dt_clf__criterion': ['gini', 'entropy']}

# Define Pipeline
dt_pipe = Pipeline([('standard_scaler', standard_scaler),
                     ('dt_clf', dt_clf)])

# Grid Search
dt_grid = GridSearchCV(dt_pipe, param_grid=param_grid, n_jobs=-1)
start_time = time.time()
dt_grid.fit(X_train, y_train.to_numpy().ravel())
end_time = time.time()
dt_time = (end_time - start_time)/60

# Get predictions
dt_test_preds = dt_grid.predict(X_test)
dt_train_preds = dt_grid.predict(X_train)

# Display Confusion Matrix
ConfusionMatrixDisplay(confusion_matrix(y_test, dt_test_preds)).plot()
plt.suptitle('Descision Treee Classifier Confusion Matrix')

# Print the classification report
print('Training Classification Report ****************')
print(classification_report(y_train, dt_train_preds))
print('Test Classification Report ********************')
print(classification_report(y_test, dt_test_preds))

# Intialize Result data frame
dt_results_df = pd.DataFrame({'Model': ['Decision Tree'],
                              'Train Time': [f'{dt_time:.3f}m'],
                              'Train Accuracy': [accuracy_score(y_train, dt_train_preds)],
                              'Train Recall': [recall_score(y_train, dt_train_preds)],
                              'Test Accuracy': [accuracy_score(y_test, dt_test_preds)],
                              'Test Recall': [recall_score(y_test, dt_test_preds)]})

results_df = pd.concat([results_df, dt_results_df])

display(results_df)
````
